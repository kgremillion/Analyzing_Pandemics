```python
#Imported libraries to read and analysis data 

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Reading the data base from WHO.org regarding the H1N1 Epidemia

n1h1_data= pd.read_csv('FluNetInteractiveReport.csv')
n1h1_data.head(50)
```

```python
#Renaming the columns for easy IDF

rename_h1n1 = n1h1_data.rename(columns = {'SDATE':'START_DATE', 'EDATE':'END_DATE','SPEC_RECEIVED_NB': 'NB_SPEC_RECEIVED','SPEC_PROCESSED_NB': 'NB_SPEC_PROCESSED','AH1': 'TYPEA H1','AH1N12009':'TYPE A H1N1','AH3':'TYPE A H2','AH5':'TYPE A H5','ANOTSUBTYPED':'TYPE A NOT SUBTYPED','INF_A':'TYPE A TOTAL','BYAMAGATA': 'TYPE B YAMAGATA','BVICTORIA':'TYPE B VICTORIA','BNOTDETERMINED':'TYPE B NOT DETERMINED','INF_B':'TYPE B TOTAL','ALL_INF':'TOTAL POS INFLUENZA','ALL_INF2':'TOTAL NEG INFLUENZA','TITLE':'ACTIVITY'})
rename_h1n1
```

```python
#Group Data by Country 

group_h1n1 = rename_h1n1.groupby(['Country'])
group_h1n1.head(0)
print(group_h1n1.describe())
```
