```python
#Imported libraries to read and analysis data 

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from itertools import groupby
%matplotlib inline
```

```python
#Reading the data base from WHO.org regarding the H1N1 Epidemia

influenza_data= pd.read_csv('Influenza Raw Data 2008-2010 Americas Region .csv')

influenza_data
```

```python
#insert a column to sum the two type of influenza analyzed 
influenza_data['Sum_Influenza_Cases'] = influenza_data['AH1']+influenza_data['AH1N12009']

influenza_data.head(5)

# df['Discounted_Price'] = df['Cost'] - (0.1 * df['Cost']) 


```

```python
#Creating the Data Frame for 2008
Data2008_flu= influenza_data.loc[influenza_data['Year'] ==2008]

Data2008_flu

# tum_line = mouse_merge_pd.loc[(mouse_merge_pd['Drug Regimen'] == "Capomulin")&(mouse_merge_pd['Mouse ID'] == "l509")][['Timepoint','Tumor Volume (mm3)']]
# new_tum_line = tum_line.reset_index()
```

```python
#replace the NaN value in the dataframe
Data2008_flu = Data2008_flu.replace(np.nan,'', regex = True)
Data2008_flu
```

```python
weather_flu_df = influenza_data[['Year','Weather','Week','Country','AH1','AH1N12009']]

weather_flu_df
```

```python
#Creating a Pivot Table 
weather_flu_df
weather_flu_df.pivot_table(index =['Year', 'Week', 'Weather'], columns='Country', values=['AH1', 'AH1N12009'], aggfunc='sum')
weather_pivot_index = weather_flu_df.set_index(['Year', 'Week', 'Weather']).sort_index()

```

```python
weather_pivot_index.plot.line()

```

```python
weather_pivot_index.plot(kind = "line", figsize=(20,15))

#formating the graph
plt.title('Relationship Outbreaks and Weather')
plt.ylabel('# of Cases')
plt.xlabel('Year / Week / Weather')
plt.grid(alpha=0.2)


plt.savefig('output_graphs/Relation of Flu vs Weather.png')
```

```python
weather_pivot_index.plot.line(figsize=(30, 15));
```

```python
weather_pivot_index.plot(subplots=True, figsize=(20, 15))

#formating the graph
plt.title('Relationship Between Type of OutBreak and Weather')
plt.ylabel('# of Cases')
plt.xlabel('Year / Week / Weather')
plt.grid(alpha=0.2)


plt.savefig('output_graphs/Relation of by type Flu vs Weather.png')
```
